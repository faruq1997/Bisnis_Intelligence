# Decision Support System (DSS)

---

## üß† **Decision Support System (DSS): Implementasi pada Kasus Fruit Glow**

### üìå Tujuan DSS di Fruit Glow
Membantu manajemen *Fruit Glow* dalam:
- Menentukan **wilayah prioritas pemasaran**.
- Mengetahui **channel distribusi** paling efektif.
- Menganalisis **kinerja tim sales**.
- Menentukan **strategi produk** berdasarkan bundling, harga, dan permintaan.
- Memberikan **rekomendasi keputusan** berbasis data.

---

## üîç **1. Fase DSS Berdasarkan Studi Kasus Fruit Glow**

### üß≠ Fase Intelijen
- **Masalah:** Penjualan di beberapa provinsi sangat rendah (Quantity <5).
- **Peluang:** Platform Shopee dan Tokopedia menunjukkan volume tinggi.
- **Kebutuhan:** Menyesuaikan strategi pemasaran & penempatan sales di wilayah tertentu.

### üèóÔ∏è Fase Desain
- Menganalisis data historis 12 bulan.
- Membandingkan performa bundling vs non-bundling.
- Menentukan kriteria keberhasilan wilayah (misalnya: penjualan >Rp5 juta/bulan).

### üß© Fase Pilihan
- Rekomendasi memperkuat promosi bundling karena memiliki AOV lebih tinggi.
- Fokus promosi Shopee dan TikTok di daerah dengan volume tinggi.
- Kirim sales baru ke wilayah "lemah" atau gunakan strategi reseller lokal.

### üöÄ Fase Implementasi
- Menyusun *campaign* bundling dengan diskon 20% untuk IG & TikTok.
- Memberikan komisi lebih tinggi kepada sales yang menjangkau provinsi low-quantity.
- Membuat *landing page* khusus bundling untuk platform paling efektif.
