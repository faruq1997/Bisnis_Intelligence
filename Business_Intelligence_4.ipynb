# Profesi Business Intelligence Analyst

Dalam artikel https://pacmann.io/blog/perbedaaan-business-intelligence-dan-business-analytics menjelaskan terkait hal dasar dari BI dan BA.

Berikut ini **ringkasan dan analogi sederhana** mengenai perbedaan antara **Business Intelligence (BI)** dan **Business Analytics (BA)**:

---

## 🔍 **Perbedaan Utama: BI vs BA**

| Aspek                     | **Business Intelligence (BI)**                                      | **Business Analytics (BA)**                                     |
|---------------------------|---------------------------------------------------------------------|------------------------------------------------------------------|
| 🔄 **Fokus Waktu**        | Masa **lalu** dan **sekarang**                                     | Masa **depan** (prediksi & proyeksi)                            |
| 🎯 **Tujuan**             | Menjawab **"Apa yang terjadi?"**                                   | Menjawab **"Mengapa bisa terjadi?"** dan **"Apa yang akan terjadi?"** |
| ⚙️ **Fungsi Utama**       | Pemantauan, pelaporan, dan pengambilan keputusan taktis            | Prediksi, optimasi, dan strategi ke depan                        |
| 🛠️ **Tools & Teknik**    | Dashboard, OLAP, reporting                                          | Statistik, machine learning, data mining                        |
| 🧠 **Sifat Analisis**     | **Deskriptif** dan **diagnostik ringan**                            | **Prediktif** dan **preskriptif**                               |

---

## 📦 **Analogi Sederhana**

Bayangkan kamu **menjalankan toko skincare** (seperti *Fruit Glow*):

- 📊 **Business Intelligence (BI)** itu seperti **membuka laporan kas harian** dan melihat:
  - "Berapa produk terjual kemarin?"
  - "Channel mana yang paling ramai bulan lalu?"

  > **BI** = Membaca *rapor bisnis* untuk tahu kinerja yang sudah terjadi.

- 🔮 **Business Analytics (BA)** itu seperti **berkonsultasi dengan analis**:
  - "Apa alasan penurunan penjualan bulan ini?"
  - "Jika tren TikTok naik, apakah lebih baik kita alihkan promosi ke sana bulan depan?"

  > **BA** = Menggunakan data untuk *memprediksi & menyarankan tindakan masa depan*.

---

## 🎯 Contoh Kasus:

| Studi Kasus *Fruit Glow*                  | Business Intelligence (BI)                  | Business Analytics (BA)                                   |
|-------------------------------------------|---------------------------------------------|------------------------------------------------------------|
| Penjualan di Jawa Barat menurun 20%       | Melihat data laporan bulan ini vs bulan lalu| Menganalisis penyebab & membuat simulasi strategi promosi |
| Produk bundling meningkat 35% di Shopee   | Menyajikan grafik tren penjualan            | Memprediksi potensi penjualan jika bundling diperluas      |

BA tidak hanya fokus pada data, namun juga identifikasi fenomena sekitar atau fenomena market. Kasus lain misalnya seorang trader saham selain perlu menganalisis grafik data juga perlu memahami keadaan global atau pasar seperti apa (kebijakan negara, kebijakan global, isu global, dsb).

Sedang Business intelligence adalah sistem yang bertujuan untuk menemukan insight dari hasil analisis data dan informasi bisnis dalam jumlah besar yang dimiliki oleh perusahaan.

Data tersebut digali dan dianalisis dengan tujuan untuk menyediakan landasan untuk perusahaan dalam menentukan keputusan bisnis, mewujudkan potensi penuh dari data yang dipunyai oleh perusahaan, dan menyediakan insight yang bernilai untuk bisnis.

### Bagaimana memahami data?

Pemahaman terhadap data atau **Literasi Data** sangat penting di era digital seperti sekarang. Literasi data bukan cuma soal bisa membaca angka di Excel, tapi tentang **memahami makna, konteks, dan cara menggunakan data untuk mengambil keputusan yang lebih baik.**

---

## 📚 **Apa itu Literasi Data?**

**Literasi Data** adalah kemampuan untuk:
1. **Membaca data** → Memahami grafik, tabel, dan laporan.
2. **Menganalisis data** → Menarik kesimpulan dari data yang tersedia.
3. **Menginterpretasikan data** → Memahami konteks & makna di balik angka.
4. **Mengomunikasikan data** → Menyampaikan informasi secara visual dan verbal.
5. **Mengambil keputusan berbasis data** → Tidak hanya berdasarkan intuisi, tetapi bukti.

---

## 🧠 **Mengapa Literasi Data itu Penting?**

- 💡 **Menghindari kesalahan dalam interpretasi** (misalnya: membaca grafik yang menyesatkan).
- 📈 **Meningkatkan akurasi pengambilan keputusan**.
- 🧩 **Menemukan insight tersembunyi** dari kumpulan data besar.
- 🧪 **Membedakan data yang kredibel dan manipulatif**.

---

## 🧰 **Cara Meningkatkan Literasi Data**

| Langkah | Penjelasan |
|--------|------------|
| 🔍 **Kenali jenis data** | Kuantitatif (angka) vs Kualitatif (narasi), data primer vs sekunder |
| 📊 **Belajar visualisasi** | Pelajari grafik batang, garis, pie, scatter plot, heatmap |
| 🗂️ **Pahami konteks data** | Siapa yang mengumpulkan, bagaimana, dan untuk apa data dibuat? |
| 🔢 **Gunakan tools sederhana** | Excel, Google Sheets, Google Data Studio, atau Python Pandas |
| 📚 **Belajar dari kasus nyata** | Gunakan studi kasus seperti *Fruit Glow* untuk memahami data penjualan |
| ✍️ **Latihan membuat insight** | Tulis kesimpulan dari data: tren, anomali, pola |

---

## 📦 **Contoh Praktis Literasi Data dari Fruit Glow**

Misalnya kamu lihat data ini:

| Bulan       | Channel   | Jumlah Terjual | Pendapatan |
|-------------|-----------|----------------|------------|
| Januari     | Shopee    | 1.200          | 36.000.000 |
| Januari     | Offline   | 150            | 4.500.000  |

🔎 **Apa yang bisa dipahami?**
- Channel Shopee sangat mendominasi → strategi pemasaran bisa fokus ke marketplace.
- Offline hanya di Jawa Barat → potensi perluasan bisa dievaluasi.

📣 **Bagaimana komunikasinya?**
> “Berdasarkan data Januari, Shopee menyumbang 89% penjualan. Ini menunjukkan potensi kuat di channel online dan peluang efisiensi biaya operasional dibanding offline.”

---

## 🎯 **Peran dan Tanggung Jawab Business Intelligence Analyst (BI Analyst)**

Business Intelligence Analyst berperan sebagai *jembatan antara data dan keputusan bisnis*. Tugas utamanya adalah **mengubah data mentah menjadi informasi strategis** yang bisa digunakan oleh manajemen untuk menentukan arah perusahaan.

---

### 1. 🔧 **Mengolah Data (Data Processing & Preparation)**

> *“Data adalah bahan mentah, tugas BI Analyst adalah menjadikannya masakan yang siap saji.”*

#### Aktivitas:
- Mengumpulkan data dari berbagai sumber: penjualan, customer service, pemasaran, keuangan, dll.
- Melakukan **data cleaning**: menghapus duplikat, mengisi nilai kosong, memperbaiki format.
- Membuat prosedur standar untuk pengumpulan data (data governance).
- Menjamin data tersimpan dan dianalisis secara **tepat dan konsisten**.

#### Contoh dari Fruit Glow:
- Menggabungkan data penjualan dari Shopee, Tokopedia, TikTok, dan offline.
- Memastikan semua tanggal penjualan berada di akhir bulan (rekapan per 28).
- Menstandarkan nama provinsi (misal: “DKI Jakarta” dan “Jakarta” disatukan).

---

### 2. 🔍 **Menafsirkan Data (Data Interpretation & Analysis)**

> *“Data tells a story. BI Analyst membantu membacanya dengan benar.”*

#### Aktivitas:
- Menemukan pola yang bisa meningkatkan kinerja bisnis (contoh: channel penjualan mana paling efektif).
- Menganalisis tren pasar atau perilaku konsumen.
- Mengembangkan **metodologi analisis** untuk membaca performa atau merespons perubahan pasar.

#### Contoh dari Fruit Glow:
- Melihat tren bahwa bundling produk lebih diminati di marketplace.
- Menemukan bahwa provinsi luar Jawa memiliki rata-rata quantity <5, butuh strategi promosi khusus.

---

### 3. 📊 **Membagikan Hasil Penemuan (Data Visualization & Reporting)**

> *“Data yang tidak dikomunikasikan dengan baik, seperti ide yang tidak pernah diutarakan.”*

#### Aktivitas:
- Membuat **visualisasi interaktif** (grafik, diagram, dashboard) untuk mempermudah pemahaman data.
- Menyusun laporan analitik dan menyampaikannya kepada manajemen.
- Menyampaikan insight yang actionable dan mudah dimengerti oleh tim non-teknis.

#### Contoh dari Fruit Glow:
- Membuat dashboard berisi:
  - Penjualan per bulan per channel.
  - Performa masing-masing sales.
  - Provinsi dengan penjualan tertinggi dan terendah.
- Menyajikan hasil dalam bentuk chart di Google Data Studio atau Python Seaborn/Plotly.

---

## 👨‍💼 Analogi: BI Analyst seperti **“Navigator Kapal Bisnis”**

- Kapten (CEO) butuh peta dan informasi arus laut.
- BI Analyst menyusun peta berdasarkan data (penjualan, pelanggan, tren).
- Tanpa BI Analyst, kapal bisa nyasar — keputusan bisnis jadi spekulatif, bukan data-driven.

---

## 📌 Kesimpulan

**Business Intelligence Analyst = Detektif Data Perusahaan**

| Tahapan | Tugas Utama |
|--------|-------------|
| 1. Mengolah Data | Membersihkan & menggabungkan data dari berbagai sumber |
| 2. Menafsirkan Data | Mencari pola, anomali, dan insight untuk keputusan bisnis |
| 3. Membagikan Penemuan | Menyampaikan hasil lewat visualisasi dan laporan yang menarik |

---

## 🧠 **Skill Wajib Seorang Business Intelligence Analyst**

Seorang **Business Intelligence Analyst** bukan hanya seseorang yang bisa membaca data. Ia adalah **penjelajah data**, **penerjemah angka**, dan **penyampai strategi bisnis** berdasarkan bukti kuat. Untuk itu, dibutuhkan berbagai keterampilan berikut:

---

### 1. 📊 **Analytical Thinking (Berpikir Analitis)**  
> *"Kemampuan membaca data seperti detektif membaca petunjuk."*

#### Fungsi:
- Mencari pola tersembunyi dalam data.
- Mengidentifikasi penyebab dari suatu tren (kenapa penjualan menurun/naik).
- Menentukan faktor-faktor kunci dalam keputusan.

#### Analogi:
Seperti dokter yang membaca gejala pasien untuk memberi diagnosa—BI Analyst membaca data untuk memberi insight bisnis.

---

### 2. 🔎 **Riset dan Investigasi Data**  
> *“Sebelum menjawab, pastikan kamu bertanya dengan benar.”*

#### Fungsi:
- Meneliti tren pasar.
- Menyusun hipotesis data: “Apakah kampanye TikTok efektif?”
- Mengembangkan pemikiran logis dan argumentatif.

#### Tips:
Bisa menggunakan Google Trends, report industri, atau wawancara pengguna.

---

### 3. 💾 **Kemampuan Tools Database**  
> *"Data tanpa alat seperti tambang emas tanpa alat gali."*

#### Tools Populer:
- **SQL** – membaca, memfilter, dan menggabungkan data dari berbagai tabel.
- **Excel/Sheets** – dasar wajib untuk pemula.
- **SPSS / SAS / Sage** – tools statistik yang digunakan dalam industri besar.

#### Kunci:
Memahami bagaimana data disimpan, diakses, dan dimodifikasi.

---

### 4. 👨‍💻 **Bahasa Pemrograman (Python / R)**  
> *"Kalau Excel itu sepeda, Python itu mobil balap."*

#### Fungsi:
- Analisis big data secara otomatis.
- Membuat visualisasi, prediksi, dan model data.
- Melakukan web scraping, data cleaning, machine learning, dll.

#### Tools:
- Python + Pandas, NumPy, Matplotlib, Seaborn.
- R untuk analisis statistik.

---

### 5. 📈 **Visualisasi Data**  
> *“Insight bagus akan percuma kalau tidak bisa disampaikan dengan jelas.”*

#### Tools Populer:
- **Tableau**
- **Microsoft Power BI**
- **Google Data Studio**

#### Fungsi:
- Menyampaikan insight ke stakeholder non-teknis.
- Membuat dashboard interaktif.
- Mengubah data menjadi **grafik, diagram, heatmap**, dll.

---

### 6. 🗣️ **Skill Komunikasi**  
> *"BI Analyst harus bisa bicara bahasa angka dan bahasa manusia."*

#### Fungsi:
- Mampu menyampaikan insight ke manajemen dan tim non-teknis.
- Membuat laporan yang jelas, padat, dan efektif.
- Bekerja lintas tim (sales, marketing, IT, finance).

---

### 7. 💼 **Pemahaman Bisnis dan Finansial**  
> *“Kalau kamu tidak paham bisnis, kamu hanya orang yang pintar Excel.”*

#### Fungsi:
- Memahami konteks: mengapa data penting untuk strategi perusahaan.
- Mengetahui terminologi bisnis: margin, ROI, CAC, CLV, dsb.
- Mampu membuat strategi berbasis data untuk pertumbuhan.

### 🎓 **Jenjang Karier dalam Business Intelligence**

---

#### 1. **BI Administrator (Entry-Level / Junior Level)**
🧠 **Peran:**
- Mengolah dan menyiapkan data untuk tim analis dan manajemen.
- Melayani permintaan data dari stakeholders dan klien internal.
- Melakukan pemantauan kompetitor dan kondisi ekonomi berdasarkan data.

🛠 **Tanggung Jawab:**
- Cleaning dan integrasi data dari berbagai sumber.
- Menjaga ketersediaan dan kualitas data harian.
- Membantu membuat report dasar dan dashboard.

---

#### 2. **BI Analyst / Senior Analyst**
📊 **Peran:**
- Menganalisis data untuk mengungkap insight yang mendalam.
- Mendukung pengambilan keputusan strategis dan operasional.

🧩 **Tanggung Jawab:**
- Membuat visualisasi data, laporan tren penjualan, operasional, dan efisiensi.
- Menggunakan tools seperti Power BI, Tableau, SQL, Python, Excel.
- Memberikan rekomendasi bisnis berdasarkan data historis dan prediksi.

---

#### 3. **Lead BI Analyst**
🧭 **Peran:**
- Memimpin tim analis dalam suatu domain (penjualan, supply chain, keuangan, dll).
- Menjadi penghubung antara analis dengan manajer dan eksekutif.

🧠 **Tanggung Jawab:**
- Menentukan standar kualitas data dan analisis.
- Memberi coaching kepada analis junior.
- Validasi dan review hasil analisis sebelum dipresentasikan ke manajemen.

---

#### 4. **Master Data Services (Data Governance)**
🔐 **Peran:**
- Menjaga konsistensi dan kualitas data lintas departemen.

⚙️ **Tanggung Jawab:**
- Menentukan definisi standar data (misalnya, satuan produk, kode wilayah).
- Mengatur lalu lintas dan akses data antar tim dan sistem.
- Menjadi penjaga “single source of truth”.

---

#### 5. **BI Developer**
👨‍💻 **Peran:**
- Membangun dan mengotomasi pipeline data dan dashboard.

🔧 **Tanggung Jawab:**
- Membuat sistem ETL (Extract, Transform, Load).
- Mendesain dan membangun database/data warehouse.
- Membuat sistem reporting otomatis.

---

#### 6. **BI Consultant**
💼 **Peran:**
- Bekerja sebagai pihak eksternal atau internal untuk memberikan solusi BI bagi perusahaan klien.

🛠 **Tanggung Jawab:**
- Mendesain arsitektur BI.
- Memberi pelatihan implementasi tools BI.
- Membantu migrasi data ke sistem BI baru.

---

#### 7. **BI Project Manager**
📅 **Peran:**
- Memimpin proyek-proyek pengembangan BI dalam organisasi.

📌 **Tanggung Jawab:**
- Menentukan scope dan timeline proyek.
- Berkoordinasi dengan tim developer, analyst, dan business user.
- Mengelola risiko dan anggaran proyek BI.

---

#### 8. **BI Manager**
👑 **Peran:**
- Memimpin keseluruhan fungsi Business Intelligence di perusahaan.

🏢 **Tanggung Jawab:**
- Menyusun roadmap BI perusahaan.
- Mengelola beberapa tim analis, developer, dan data engineer.
- Menjaga kolaborasi antara BI dan divisi bisnis lainnya (marketing, finance, sales, dll).
- Memastikan data menjadi aset strategis untuk pertumbuhan bisnis.

---
